# **The Dot Product**

```python
a = np.array([1, 2])
b = np.array([3, 4])
```

```python
dot = 0
for e, f in zip(a, b):
  dot += e * f
dot
```

```python
dot = 0
for i in range(len(a)):
  dot += a[i] * b[i]
dot
```

```python
a * b
np.sum(a*b)
```

```python
(a*b).sum()
```

```python
np.dot(a,b)
```

```python
a.dot(b)
```

```python
a @ b
```

```python
# Other formula
amag = np.sqrt((a*a).sum())
# or it is the same
amag = np.linalg.norm(a)
```

```python
cosangle = a.dot(b) / (np.linalg.norm(a) * np.linalg.norm(b))
cosangle
```

```python
angle = np.arccos(cosangle)
angle
```

```python
B = np.array([[1,2,3], [4,5,6]])
B
```

```python
A.dot(B)
```

```python
# Illegal - doesn't match dimensions
# A.dot(B.T)
```

```python
np.linalg.det(A)
```

```python
np.linalg.inv(A)
```

```python
np.trace(A)
```

```python
np.diag(A)
```

```python
np.diag([1,4])
```

```python
Lam, V = np.linalg.eig(A)
```

```python
V[:, 0] * Lam[0] == A @ V[:, 0]
```

```python
V[:, 0] * Lam[0], A @ V[:, 0]
```

```python
# Use np.allclose instead of ==
np.allclose(V[:, 0] * Lam[0], A @ V[:, 0])
```

```python
np.allclose(V @ np.diag(Lam), A @ V)
```

# **Solving Linear Systems**

ex: 
* x1 + x2 = 2200
* 1.5x1 + 4x2 = 5050

```python
A = np.array([[1,1], [1.5, 4]])
b = np.array([2200, 5050])
```

```python
np.linalg.solve(A, b)
```
