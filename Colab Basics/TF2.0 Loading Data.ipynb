```python
# Check the data
import pandas as pd
df = pd.read_csv("arrhythmia.data", header=None)
```

```python
# Data has many columns -> so we will just the first few and rename them (as per the UCI documentation)
data = df[[0,1,2,3,4,5]]
data.columns = ['age', 'sex', 'height', 'weight', 'QRS Duration', 'P-R Interval']
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 15] # Make the plot bigger so the plots do not overlap
data.hist(); # Use a semicolot to supress the return value
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(data);
```
