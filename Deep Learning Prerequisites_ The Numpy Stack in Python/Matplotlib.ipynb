# **Line Chart**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 1000)
```

```python
y = np.sin(x) + 0.2 * x
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my plot');
```

# **Scatterplot**

```python
X = np.random.randn(100, 2)
plt.scatter(X[:,0], X[:, 1])
```

```python
X = np.random.randn(200, 2)
X[:50] += 3
Y = np.zeros(200)
Y[:50] = 1
plt.scatter(X[:, 0], X[:,1], c = Y)
```

# **Excercise**

```python
X1 = np.concatenate([np.random.uniform(low = 0, high = 1, size = 100), np.random.uniform(low = -1, high = 0, size = 100)])
y1 = np.concatenate([np.random.uniform(low = -1, high = 0, size = 100), np.random.uniform(low = 0, high = 1, size = 100)])

X2 = np.concatenate([np.random.uniform(low = 0, high = 1, size = 100), np.random.uniform(low = -1, high = 0, size = 100)])
y2 = np.concatenate([np.random.uniform(low = 0, high = 1, size = 100), np.random.uniform(low = -1, high = 0, size = 100)])

plt.scatter(X1, y1, c='red')
plt.scatter(X2, y2, c='blue')
plt.show
```
